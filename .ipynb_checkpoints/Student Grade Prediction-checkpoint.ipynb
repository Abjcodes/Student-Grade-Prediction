```python
data = df[['G3']]
pyplot.figure(figsize=(10,10))
g = sns.heatmap(df.corr(),annot=True,cmap="RdYlGn")
```

G1,G2,Medu,Fedu,studytime,famrel,absences are the most important.
Eventhough failures is showing low correlation, it is known that failures play an important role so it is also included

```python
data2 = df[['absences', 'G1', 'G2','studytime', 'famrel', 'Medu', 'Fedu', 'failures']]
```

```python
x = np.array(data2)

y = np.array(data)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size= 0.1)
```

```python
best_score = 0
```

```python
for i in range(50):

    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size= 0.1)

    linear = linear_model.LinearRegression()

    linear.fit(x_train,y_train)
    accuracy = linear.score(x_test, y_test)

    #print(accuracy)
    if accuracy>best_score:
        best_score = accuracy
        with open("grademodel.pickle", "wb")as f:
            pickle.dump(linear, f)
print("best outcome is " + str(best_score))
```

```python
 pickle_in = open("grademodel.pickle", "rb")
 linear = pickle.load(pickle_in)
```

```python
#Prediction
predictions = linear.predict(x_test)
for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```

```python
np.sqrt(metrics.mean_squared_error(y_test[x], predictions[x]))
```
